```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
iris = sns.load_dataset('iris')
```

```python
print(iris.head())
```

```python
for column in iris.columns[:-1]:  # Exclude the 'species' column
    sns.kdeplot(iris[column], fill=True)
    plt.title(f'Distribution of {column}')
    plt.show()
```

```python
iris.hist(figsize=(10, 8))
plt.suptitle('Histograms of Iris Features')
plt.show()
```

```python
sns.pairplot(iris, hue='species')
plt.suptitle('Pairplot of Iris Dataset', y=1.02)
plt.show()
```

```python
for column in iris.columns[:-1]:  # Exclude the 'species' column
    sns.boxplot(x='species', y=column, data=iris)
    plt.title(f'Boxplot of {column} by Species')
    plt.show()
```

```python
corr_matrix = iris.drop(columns='species').corr()
```

```python
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title('Heatmap of Feature Correlations')
plt.show()
```
